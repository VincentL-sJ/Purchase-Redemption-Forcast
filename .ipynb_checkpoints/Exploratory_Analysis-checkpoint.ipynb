```python
user_balance_table = pd.read_csv('data/user_balance_table.csv')
```

```python
user_balance_table[user_balance_table.user_id == 6].sort_values('report_date')
```

```python
user_balance_table
```

```python
user_balance_table.report_date = pd.to_datetime(user_balance_table['report_date'], format='%Y%m%d', errors='ignore')
```

```python
grouped = user_balance_table.groupby(['user_id'])['report_date'].agg(['min', 'max'])
```

```python
grouped['user_type'] = np.where(grouped[max]-grouped[min] != '0 days', 'old', 'new')
```

```python
user_type = grouped[grouped['user_type'] == 'new'].reset_index()
```

```python
user_type
```

```python
new_user = user_balance_table.merge(user_type, on = 'user_id',how = 'inner')
```
